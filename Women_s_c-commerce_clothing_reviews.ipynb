```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

df = pd.read_csv("Women_s_E-Commerce_Clothing_Reviews_1594_1.csv", sep=";")

df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('.', '_')
df.head()
```

```python
def veri_seti_ozeti(df):
    print("###########################################################################################################################")
    print(df.info())
    print("#############################################-Her bir sütundaki NaN değer sayısı-#############################################")
    print(df.isnull().sum())
    print("###########################################################################################################################")
    for column in df.columns:
        print(f"Sütun: {column}")
        print(f"Unique değerler: {df[column].unique()}")
        print(f"Unique değer sayısı: {len(df[column].unique())}")
        print("###########################################################################################################################")
veri_seti_ozeti(df)
```

```python
import nltk
import re
import string

nltk.download('stopwords')

def clean_text(text):
    text = text.lower()  
    text = re.sub(r'\[.*?\]', '', text) 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  
    text = re.sub(r'<.*?>+', '', text) 
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  
    text = re.sub(r'\n', '', text)  
    text = re.sub(r'\w*\d\w*', '', text)  
    return text


df['review_text'] = df['review_text'].astype(str).apply(clean_text)

df.head()
```

```python
from nltk.corpus import stopwords

sw = stopwords.words('english')
df['review_text'] = df['review_text'].apply(lambda x: " ".join(x for x in str(x).split() if x not in sw))
```

```python
all_words = ' '.join(df['review_text']).split()
word_counts = pd.Series(all_words).value_counts()

filtered_word_counts = word_counts[word_counts <= 1]

df['review_text'] = df['review_text'].apply(lambda x: " ".join(x for x in x.split() if x not in filtered_word_counts))
```

```python
import nltk
from textblob import Word

nltk.download('wordnet')
df['review_text'] = df['review_text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()


df['negative'] = df['review_text'].apply(lambda txt: sia.polarity_scores(txt)).apply(lambda txt: txt['neg'])
df['neutral'] = df['review_text'].apply(lambda txt: sia.polarity_scores(txt)).apply(lambda txt: txt['neu'])
df['positive'] = df['review_text'].apply(lambda txt: sia.polarity_scores(txt)).apply(lambda txt: txt['pos'])
df['compound'] = df['review_text'].apply(lambda txt: sia.polarity_scores(txt)).apply(lambda txt: txt['compound'])
df["sentiment_label"] = df["review_text"].apply(lambda x: "pos" if sia.polarity_scores(x)["compound"] >= 0.5 else "neg")

df.head()
```

```python
def polarity_score(compound):
    if compound > 0.05:
        return "positive"
    elif compound < -0.5:
        return "negative"
    elif compound >= -0.05 and compound < 0.05:
        return "neutral"
    

df['sentiment'] = df['compound'].apply(lambda val: polarity_score(val))
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

df["sentiment_label"] = LabelEncoder().fit_transform(df["sentiment_label"])
```

```python
X = df["review_text"]
y = df["sentiment_label"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)


tf_idf_word_vectorizer = TfidfVectorizer()
X_train_tf_idf_word = tf_idf_word_vectorizer.fit_transform(X_train)
X_test_tf_idf_word = tf_idf_word_vectorizer.transform(X_test)

tf_idf_ngram_vectorizer = TfidfVectorizer(ngram_range=(2, 3))
X_train_tf_idf_ngram = tf_idf_ngram_vectorizer.fit_transform(X_train)
X_test_tf_idf_ngram = tf_idf_ngram_vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

# Count Vectors
print("Count Vectors")
log_model = LogisticRegression().fit(X_train_count, y_train)
y_pred_count = log_model.predict(X_test_count)
print(classification_report(y_pred_count, y_test))
print(cross_val_score(LogisticRegression(), CountVectorizer().fit_transform(X), y, scoring="accuracy", cv=3).mean())

# TF-IDF Word-Level
print("TF-IDF Word-Level")
log_model = LogisticRegression().fit(X_train_tf_idf_word, y_train)
y_pred_tf_idf_word = log_model.predict(X_test_tf_idf_word)
print(classification_report(y_pred_tf_idf_word, y_test))
print(cross_val_score(LogisticRegression(), TfidfVectorizer().fit_transform(X), y, scoring="accuracy", cv=3).mean())

# TF-IDF N-GRAM
print("TF-IDF N-GRAM")
log_model = LogisticRegression().fit(X_train_tf_idf_ngram, y_train)
y_pred_tf_idf_ngram = log_model.predict(X_test_tf_idf_ngram)
print(classification_report(y_pred_tf_idf_ngram, y_test))
print(cross_val_score(LogisticRegression(), TfidfVectorizer(ngram_range=(2, 3)).fit_transform(X), y, scoring="accuracy", cv=3).mean())
```

Count Vectors:

Precision, recall ve f1-score değerleri sınıf 0 için 0.68, 0.79, 0.73; sınıf 1 için ise 0.97, 0.95, 0.96 olarak yüksektir.
Bu, modelin sınıf 1'i (olumlu yorumları) iyi tahmin ettiğini ancak sınıf 0'ı (olumsuz yorumları) daha az doğru tahmin ettiğini gösterir.
Accuracy %93 olarak yüksek, yani genel olarak model iyi performans gösteriyor.
Cross-validation score (0.925) da yüksek, modelin genel geçerliliğini destekliyor.

TF-IDF Word-Level:

Precision, recall ve f1-score değerleri sınıf 0 için 0.43, 0.94, 0.59; sınıf 1 için ise 1.00, 0.91, 0.95 olarak görünmektedir.
Sınıf 0 için precision düşüktür, yani sınıf 0 olarak tahmin edilenlerin çoğu gerçekten sınıf 0 değildir.
Recall sınıf 1 için düşüktür, yani gerçek sınıf 1'lerin çoğu tespit edilmemiştir.
Accuracy %92 olarak yüksek, ancak Count Vectors'a göre daha düşüktür.
Cross-validation score (0.910) da yine Count Vectors'a göre düşüktür, bu da genel geçerliliğin Count Vectors'a kıyasla daha düşük olduğunu gösterir.

TF-IDF N-GRAM:

Precision, recall ve f1-score değerleri düşüktür, sınıf 0 ve sınıf 1 için de benzer düşük değerler görülmektedir.
Model, N-GRAM özelliklerini kullanarak daha kötü performans göstermektedir.
Accuracy %86 olarak düşüktür, modelin genel olarak kötü performans gösterdiğini gösterir.
Cross-validation score (0.851) da düşüktür, bu da modelin genel geçerliliğinin zayıf olduğunu doğrular.
Sonuç olarak, Count Vectors yöntemi diğerlerine kıyasla daha iyi bir performans sergilemiştir. Daha yüksek precision, recall, f1-score ve accuracy değerleri ile daha güvenilir tahminler yapabilmektedir. Diğer yöntemlerin performansı ise belirli metin verileri setine ve sınıflandırma görevine bağlı olarak değişiklik gösterebilir.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Count Vectors
print("Count Vectors")
rf_model = RandomForestClassifier().fit(X_train_count, y_train)
y_pred_count = rf_model.predict(X_test_count)
print(classification_report(y_pred_count, y_test))
print(cross_val_score(RandomForestClassifier(), CountVectorizer().fit_transform(X), y, scoring="accuracy", cv=3).mean())


# TF-IDF Word-Level
print("TF-IDF Word-Level")
rf_model = RandomForestClassifier().fit(X_train_tf_idf_word, y_train)
y_pred_tf_idf_word = rf_model.predict(X_test_count)
print(classification_report(y_pred_tf_idf_word, y_test))
print(cross_val_score(LogisticRegression(), TfidfVectorizer().fit_transform(X), y, scoring="accuracy", cv=3).mean())


# TF-IDF N-GRAM
print("TF-IDF N-GRAM")
rf_model = RandomForestClassifier().fit(X_train_tf_idf_ngram, y_train)
y_pred_tf_idf_ngram = rf_model.predict(X_test_tf_idf_ngram)
print(classification_report(y_pred_tf_idf_ngram, y_test))
print(cross_val_score(RandomForestClassifier(), TfidfVectorizer(ngram_range=(2, 3)).fit_transform(X), y, scoring="accuracy", cv=3).mean())
```
